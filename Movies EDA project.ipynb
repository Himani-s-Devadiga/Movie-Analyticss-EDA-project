```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('movies.csv')
df
```

```python
#Handling missing values
df.isnull().sum()
```

```python
df.fillna({'director':'Unknown', 'cast':'Unknown', 'country':'Unknown'}, inplace=True)
```

```python
df.columns
```

```python
import re

def extract_year(date):
    if isinstance(date, str):
        match = re.search(r'\d{4}', date)
        if match:
            return int(match.group())
    return None

df['release_year'] = df['release_date'].apply(extract_year)
```

```python
print(df['release_year'].head(20))
print(df['release_year'].isnull().sum(), "missing values")
```

```python
if df['release_year'].dropna().shape[0] > 0:
    df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(12,4))
    plt.title('Number of Titles Released Per Year')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.show()
else:
    print("No valid release_year values to plot.")
```

```python
#Movies vs TV Shows
df['type'] = df['release_date'].apply(lambda x: 'TV Show' if isinstance(x, str) and 'TV Series' in x else 'Movie')

sns.countplot(data=df, x='type')
plt.title('Movies vs TV Shows')
plt.show()
```

```python
#Top Genres
from collections import Counter

genre_list = df['genres'].str.split('|').sum()
top_genres = pd.Series(genre_list).value_counts().head(10)
top_genres.plot(kind='barh', title='Top 10 Genres')
plt.show()
```

```python
#IMDB Ratings Distribution
sns.histplot(df['imdb_rating'], bins=20, kde=True)
plt.title('Distribution of IMDB Ratings')
plt.xlabel('Rating')
plt.show()
```

```python
#Runtime Distribution
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')
sns.histplot(df['runtime'].dropna(), bins=20)
plt.title('Movie Runtime Distribution')
plt.xlabel('Runtime (minutes)')
plt.show()
```

```python
#Top Actors
actor_list = df['actors'].dropna().str.split('|').sum()
top_actors = pd.Series(actor_list).value_counts().head(10)
top_actors.plot(kind='barh', title='Top 10 Actors')
plt.show()

```

```python
# Explode genres to separate rows
df_genres = df.copy()
df_genres = df_genres.assign(genres=df_genres['genres'].str.split('|')).explode('genres')

sns.boxplot(data=df_genres, x='genres', y='imdb_rating')
plt.xticks(rotation=45)
plt.title('IMDB Rating Distribution by Genre')
plt.show()
```

# Key Insights from Movie Analytics EDA
Drama is the most common genre

From the genre analysis, Drama appears in the largest number of movies, followed by Action and Comedy.

Movies vastly outnumber TV shows

Most entries in the dataset are movies, with only a few TV shows like Humsafar.

Highest-rated actors

Actors like Vicky Kaushal, Ajay Devgn, and Anupam Kher appear frequently in movies with high IMDB ratings.

Top-awarded movies

Movies like Uri: The Surgical Strike and Evening Shadows have won the highest number of awards (wins + nominations).

Most popular movies by votes

Movies with the largest number of IMDB votes include Uri: The Surgical Strike, indicating high audience engagement.

Distribution of IMDB ratings

Most movies have ratings between 6.0 and 8.0, showing that average audience reception is moderately high.

Runtime trends

Most movies have runtimes between 120–150 minutes, which is the standard for commercial films.

Relationship between votes and rating

Scatter plots indicate that movies with higher votes often have higher ratings, but there are exceptions — some niche films have high ratings but fewer votes.

Genre-wise rating patterns

Action and Drama movies tend to have higher median ratings compared to Comedy and Romance genres.

Popular actors appear across genres

Frequent actors often act in multiple genres, showing versatility and consistent audience interest.
